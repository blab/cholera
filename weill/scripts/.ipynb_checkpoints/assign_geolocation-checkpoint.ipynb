```python
df["region"] = np.nan
df["region"][df.Country.isin(north_america)] = "North America"
df['region'][df.Country.isin(europe)] = "Europe"
df['region'][df.Country.isin(oceania)] = "Oceania"
df['region'][df.Country.isin(africa)] = "Africa"
df['region'][df.Country.isin(asia)] = "Asia"
df['region'][df.Country.isin(south_america)] = "South America"
df['region'][df.Country.isin(unknown)] = "Unknown" #######
```

```python
df.region = df.region.str.replace(' ', '_')
df.Country = df.Country.str.replace(' ', '_')
```
